## Object Thinking in Python

In Python, everything is an object. Variables do not *contain* objects in the way a box contains items; instead, a variable is a **reference** (or label) to an object.

The next small example is meant to make that visible.

````python

# A quick warm-up: variables as references

basket_a = ["apple", "orange"]
basket_b = basket_a

basket_b.append("pear")

print("basket_a:", basket_a)
print("basket_b:", basket_b)
print("Same object?", basket_a is basket_b)
````

### Formative Exercise 1

You are helping a small campus library model books as objects.

Complete the `Book` class so that:

1. the constructor stores `title`, `author`, and `available`
2. the method `borrow()` changes the book to unavailable, but only if it is currently available
3. the method `return_book()` marks it as available again
4. the method `describe()` returns a readable sentence such as  
   `"Berserk by Kentaru Miura (available)"`

After that, create one `Book` object and test the methods.

**Question:** Which data belongs inside the object, and which actions should be methods of the object itself?

````python

# Exercise 1

class Book:
    def __init__(self, title, author, available=True):
        self.title = title
        self.author = author
        self.available = available

    def borrow(self):
        if self.available:
            self.available = False
    
    def return_book(self):
        self.available = True
    
    def describe(self):
        status = "available" if self.available else "unavailable" 
        return f"{self.title} by {self.author} ({status})" 
    
        

b1 = Book("Berserk", "Kentaru Miura")
print(b1.describe())

b1.borrow()
print(b1.describe())

b1.return_book()
print(b1.describe())
````

## Classes, Methods, and Constructors

A class becomes useful when it combines **state** (attributes) and **behavior** (methods).  
The classic `Point` example is simple, but it clearly shows how objects can model something meaningful.

### Formative Exercise 2

Implement a class `Point` that represents a point in 2D space.

Your class should include:

- a constructor `__init__(x, y)`
- a method `move(x, y)` that updates the coordinates
- a method `reset()` that moves the point to `(0, 0)`
- a method `calculate_distance(other)` that returns the Euclidean distance to another point

Then test the class by:

1. creating two points
2. printing the distance between them
3. moving one point
4. printing the new distance

**Hint:** use `math.hypot()`.

````python

# Exercise 2

import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def move(self, x, y):
        self.x = x
        self.y = y
    
    def reset(self):
        self.x = 0
        self.y = 0
    
    def calculate_distance(self, other):
        return math.hypot(self.x - other.x, self.y - other.y)


point1 = Point(0, 0)
point2 = Point(3, 4)

print("Distance: ", point1.calculate_distance(point2))

point1.move(5, 0)
print("New Distance: ", point1.calculate_distance(point2))
````

## Modules and Packages

As programs grow, we do not want everything in one cell or one file.  
Python uses **modules** (files) and **packages** (directories of modules) to keep code organized.

In a notebook, we can simulate this with `%%writefile`.

````python

# Setup for the next exercise: we need to run this cell first

from pathlib import Path

Path("campus_shop").mkdir(exist_ok=True)

with open("campus_shop/__init__.py", "w", encoding="utf-8") as f:
    f.write("# package marker\n")

with open("campus_shop/products.py", "w", encoding="utf-8") as f:
    f.write(
        "class Product:\n"
        "    def __init__(self, name, price):\n"
        "        self.name = name\n"
        "        self.price = price\n"
        "\n"
        "    def label(self):\n"
        "        return f'{self.name} (€{self.price})'\n"
    )

print("Package files created.")
````

## Inheritance and Abstract Classes

Inheritance is useful when several classes share a common structure, but differ in specific behavior.  
A good design question is: **What should be shared by all subclasses, and what must each subclass decide for itself?**

### Formative Exercise 4

A campus game club wants to model chess pieces.

Complete the code below so that:

1. `Piece` is an abstract class
2. `Piece` stores `color`
3. `Piece` requires a method `move()`
4. `Knight` and `Bishop` inherit from `Piece`
5. each subclass provides its own implementation of `move()`

Finally, create one knight and one bishop, and call their `move()` methods.

**Design question:** Why is it better to make `move()` abstract here instead of giving every piece the same default behavior?

````python

# Exercise 4

import abc

class Piece(abc.ABC):
    def __init__(self, color):
        self.color = color
    
    @abc.abstractmethod
    def move(self):
        pass

class Knight(Piece):
    def move(self):
        print(f"{self.color} knight moves in an L-shape")


class Bishop(Piece):
    def move(self):
        print(f"{self.color} bishop moves diagonally")

k  = Knight("white")
b = Bishop("black")

k.move()
b.move()
````

## Extending Built-ins and Using `super()`

Python lets us inherit from built-in classes such as `list`. This is powerful, but it should be used with care and with a clear purpose.

The next exercise wants you to extend a list in a way that feels meaningful rather than decorative :-)

### Formative Exercise 5

Create a class called `TaskList` that behaves like a `list` of tasks, with a small twist:

- When a new task is added, it should display a message:  
  `"Adding task: <task>"`
- The task should still be stored as usual (just like a normal list would do)  
  *Hint: use `super()` if needed*
- Add a method `pending_count()` that returns the number of tasks currently stored

Then, test your class by adding at least three tasks.

After running your code, also print `TaskList.mro()` and observe the order in which Python resolves methods.

````python

# Exercise 5

class TaskList(list):
    def append(self, task):
        print(f"Adding task: {task}")
        super().append(task)

    def pending_count(self):
        return len(self)

# Test your class below
tasks = TaskList()
tasks.append("Finish OOP notebook")
tasks.append("Review inheritance notebook")
tasks.append("Edit Point class")

print("Pending Tasks", tasks.pending_count())
print("Current tasks:", tasks)

print("MRO:", TaskList.mro())
````
