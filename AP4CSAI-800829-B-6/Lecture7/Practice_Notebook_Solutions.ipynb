## Encapsulation: controlling access to data

Python does not strictly prevent access to attributes in the same way as some other languages.  
Instead, it relies heavily on naming conventions:

- `name` → public
- `_name` → protected by convention
- `__name` → private using name mangling

Encapsulation helps us protect object state and control how data is changed.

### Formative Exercise 1

Create a class called `StudentProfile`.

Each student should have:

- `name` as a public attribute
- `program` as a protected attribute
- `grade` as a private attribute

Add two methods:

1. `get_grade()` returns the grade
2. `set_grade(new_grade)` updates the grade only if it is between `0` and `10`

Then create one object and test the methods.

````python

# Exercise 1
class StudentProfile:
    def __init__(self, name, program, grade):
        self.name = name # public
        self._program = program # protected
        self.__grade = grade # private

    def get_grade(self):
        return f"grade is: {self.__grade}"
    
    def set_grade(self, new_grade):
        self.__grade = new_grade

# Object Creation
s1 = StudentProfile("John", "BSc CSAI", 8.5)
print(s1.name)
print(s1._program) # accessible but not recommended
````

## Type Checking: the less flexible design

Sometimes we write code that checks the type of every object before deciding what to do.  
This can work for small programs, but it becomes hard to maintain when we add more classes.

## Polymorphism: same method, different behavior

Polymorphism lets us call the same method on different objects.  
Instead of asking *what type is this object?*, we ask *can this object do the behavior I need?*

## Exceptions: stopping invalid operations

In OOP, objects should not be allowed to enter invalid states.  
Exceptions help us stop invalid operations clearly instead of silently ignoring problems.

### Formative Exercise 9

Create a class called `BankAccount`.

It should have:

- private attribute `__balance`
- method `get_balance()`
- method `withdraw(amount)`

If the user tries to withdraw more than the balance, raise:

```python
ValueError("Insufficient funds")
```

Then test the class with a valid withdrawal and an invalid withdrawal.

````python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance
    
    def withdraw(self, amount):
        if amount>self.__balance:
            raise ValueError("Insufficient funds!")
        else:
            self.__balance -=amount

b = BankAccount(1000)
print(f"balance is: {b.get_balance()}")
b.withdraw(1100)
````

### Formative Exercise 11

Define your own exception class called `InsufficientFundsError`.

It should inherit from `Exception`.

The error message should include:

- the current balance
- the attempted withdrawal amount

Then update `BankAccount` so that it raises this custom exception instead of `ValueError`.

````python
class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        super().__init__(f"Cannot withdraw {amount} from balance {balance}")

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def withdraw(self, amount):
        if amount > self.__balance:
            raise InsufficientFundsError(self.__balance, amount)
        self.__balance -= amount


account = BankAccount(100)

try:
    account.withdraw(150)
except InsufficientFundsError as e:
    print("Transaction failed:", e)

print("Balance:", account.get_balance())
````

## Flyweight Pattern: sharing repeated data

The Flyweight pattern reduces memory usage by sharing common data between many similar objects.

In this notebook, we use a tokenizer example:

- intrinsic state: token text, token id, token length
- extrinsic state: position in the sentence

### Formative Exercise 12

Create a class called `TokenFlyweight`.

It should store only the shared token data:

- `text`
- `token_id`
- `length`

Then create two token objects manually and print their attributes.

````python
# Exercise 12

class TokenFlyweight:
    def __init__(self, text, token_id):
        self.text = text
        self.token_id = token_id
        self.length = len(text)


token1 = TokenFlyweight("code", 1)
token2 = TokenFlyweight("after", 2)

print(token1.text, token1.token_id, token1.length)
print(token2.text, token2.token_id, token2.length)
````

### Formative Exercise 13

Create a `TokenFactory`.

The factory should:

- keep a dictionary called `_tokens`
- keep a counter called `_next_id`
- have a method `get_token(text)`

Inside `get_token(text)`:

1. convert the text to lowercase
2. if the token does not exist, create a new `TokenFlyweight`
3. if the token already exists, reuse it
4. return the shared token object

````python
# Exercise 13

class TokenFactory:
    def __init__(self):
        self._tokens = {}
        self._next_id = 1

    def get_token(self, text):
        key = text.lower()

        if key not in self._tokens:
            print(f"Creating shared token: {key}")
            self._tokens[key] = TokenFlyweight(key, self._next_id)
            self._next_id += 1
        else:
            print(f"Reusing shared token: {key}")

        return self._tokens[key]


factory = TokenFactory()

t1 = factory.get_token("code")
t2 = factory.get_token("after")
t3 = factory.get_token("code")

print(t1 is t3)
````

### Formative Exercise 14

Create a class called `TokenOccurrence`.

This class should store the unique/extrinsic information:

- `token`
- `position`

Then tokenize the sentence:

```python
"code after code I write code to fix code"
```

For each word:

1. ask the factory for the shared token
2. create a `TokenOccurrence`
3. store it in a list

Finally, print each occurrence in this format:

```python
position=0, text=code, id=1
```

````python
# Exercise 14

class TokenOccurrence:
    def __init__(self, token, position):
        self.token = token
        self.position = position


sentence = "code after code I write code to fix code"
words = sentence.split()

factory = TokenFactory()
occurrences = []

for position, word in enumerate(words):
    shared_token = factory.get_token(word)
    occurrence = TokenOccurrence(shared_token, position)
    occurrences.append(occurrence)

print()
for occurrence in occurrences:
    print(
        f"position={occurrence.position}, "
        f"text={occurrence.token.text}, "
        f"id={occurrence.token.token_id}"
    )

print()
print("Number of token occurrences:", len(occurrences))
print("Number of shared token objects:", len(factory._tokens))
````

## Factory Pattern: centralizing object creation

The Factory pattern creates objects without exposing object creation logic everywhere in the code.  
Instead of writing many `if` / `elif` blocks throughout a program, we place that logic in one factory.

### Formative Exercise 15

Create three simple NLP task classes:

- `SentimentAnalyzer`
- `SpamDetectorModel`
- `TopicClassifier`

Each class should have a method called `predict(text)`.

Return simple strings such as:

- `"positive"`
- `"not spam"`
- `"education"`

````python
# Exercise 15

class SentimentAnalyzer:
    def predict(self, text):
        return "positive"


class SpamDetectorModel:
    def predict(self, text):
        return "not spam"


class TopicClassifier:
    def predict(self, text):
        return "education"


sentiment = SentimentAnalyzer()
spam = SpamDetectorModel()
topic = TopicClassifier()

print(sentiment.predict("This course is useful."))
print(spam.predict("Please read the lecture notes."))
print(topic.predict("Today we learn design patterns."))
````

### Formative Exercise 16

Create a class called `NLPFactory`.

It should have a static method called `create(task)`.

The method should return:

- `SentimentAnalyzer()` if task is `"sentiment"`
- `SpamDetectorModel()` if task is `"spam"`
- `TopicClassifier()` if task is `"topic"`

If the task is unknown, raise:

```python
ValueError("Unknown task")
```

Then create one model using the factory and call `predict()`.

````python
# Exercise 16

class NLPFactory:
    @staticmethod
    def create(task):
        if task == "sentiment":
            return SentimentAnalyzer()
        elif task == "spam":
            return SpamDetectorModel()
        elif task == "topic":
            return TopicClassifier()
        else:
            raise ValueError("Unknown task")


model = NLPFactory.create("sentiment")
print(model.predict("This notebook is helpful."))

model = NLPFactory.create("topic")
print(model.predict("Advanced OOP and design patterns"))
````

### Formative Exercise 17

Use the factory in a small application.

For each task:

3. if the task is unknown, catch the error and print a friendly message

````python
# Exercise 17

tasks = ["sentiment", "spam", "topic", "translation"]

for task in tasks:
    try:
        model = NLPFactory.create(task)
        print(task, "->", model.predict("Example input text"))
    except ValueError as e:
        print(task, "-> cannot create model:", e)
````
