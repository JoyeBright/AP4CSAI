## Encapsulation: controlling access to data

Python does not strictly prevent access to attributes in the same way as some other languages.  
Instead, it relies heavily on naming conventions:

- `name` → public
- `_name` → protected by convention
- `__name` → private using name mangling

Encapsulation helps us protect object state and control how data is changed.
